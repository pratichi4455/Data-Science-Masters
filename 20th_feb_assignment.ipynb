Q1. Explain GET and POST methods.

Answer:

- GET
The Get method is mainly used on the client (Browser) side to send a request to a specified server to get certain data or resources. Using this method the server should only let us receive the data and not change its state. Hence it is only used to view something and not to change it. The Get method is one of the most used HTTP methods. The request parameter of the get method is appended to the URL. Get request is better for the data which does not need to be secure.

- POST
The Post method is mainly used at the client (Browser) side to send data to a Specified server in order to create or rewrite a particular resource/data. This data sent to the server is stored in the request body of the HTTP request. Post method eventually leads to the creation of a new resource or updating an existing one. Due to this dynamic use, it is one of the most used HTTP methods. It is not one of the most secure methods because the data that is been sent is included in the body of the request and not in the URL. Post request is better for the data which needs to be secure

Q3. Why is redirect() used in Flask?

Answer:

In Flask, the redirect() function is used to redirect the client to a different URL. This function is often employed when you want to redirect users to a different page after performing a certain action or in response to a specific request.

Example:

from flask import Flask, redirect

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage'

@app.route('/redirect_example')
def redirect_example():
    # Redirect to the homepage
    return redirect(url_for('index'))

if __name__ == '__main__':
    app.run(debug=True)


Q4. What are templates in Flask? Why is the render_template() function used?

Answer:

In Flask, templates are files containing static or dynamic content that can be rendered by the application and returned to the client's web browser. Templates typically contain HTML markup with placeholders for dynamic data, allowing Flask to inject data into the templates before sending the final HTML response to the client.

The render_template() function in Flask is used to render a template and return it as a response to the client's request. It takes the name of the template file as its first argument and optional keyword arguments representing the variables to be passed to the template.
